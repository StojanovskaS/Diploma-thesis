Време на извршување на внес на податоци

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
details_insert = {
    'Amazon DynamoDB' : [11337],
    'Apache Cassandra' : [16.329],
    'MongoDB':[1320],
    'Neo4j':[3.44]
}
data_insert=pd.DataFrame(details_insert)
```

```python
details_insert2 = {
    'Amazon DynamoDB' : ['11337s'],
    'Apache Cassandra' : ['16.329s'],
    'MongoDB':['1320s'],
    'Neo4j':['3.44s']
}
data_insert2=pd.DataFrame(details_insert2)
```

```python
data_insert.style.applymap(lambda x: f"color: {'red' if isinstance(x,str) else 'red'}")
```

```python
cell_hover = {
    "selector": "td:hover",
    "props": [("background-color", "#FFFFE0")]
}
index_names = {
    "selector": ".index_name",
    "props": "font-style: italic; color: darkgrey; font-weight:normal;"
}
headers = {
    "selector": "th:not(.index_name)",
    "props": "background-color: #000080; color: white;"
}

data_insert2.style.set_table_styles([cell_hover, index_names, headers])
```

```python
fig = plt.figure()
data_insert.plot.bar(rot=0,figsize=(20,20))
plt.title("Време потребно за внес на податочното множество во секоја од базите во секунди", y=1.013, fontsize=22)
plt.xlabel("0-Amazon DynamoDB, 1-Apache Cassandra, 2-MongoDB, 3-Neo4j", labelpad=16)
plt.ylabel("Query Time", labelpad=16);
fig.show()
```

```python
data_insert.plot(style=['x','o','^','*'])
```
